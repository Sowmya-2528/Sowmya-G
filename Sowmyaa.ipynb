```python

import matplotlib.pyplot as plt

# Sample movie database
movies = [
    {"title": "Inception", "genres": ["Action", "Sci-Fi", "Thriller"]},
    {"title": "The Notebook", "genres": ["Romance", "Drama"]},
    {"title": "Avengers: Endgame", "genres": ["Action", "Superhero", "Adventure"]},
    {"title": "Interstellar", "genres": ["Sci-Fi", "Drama", "Adventure"]},
    {"title": "La La Land", "genres": ["Romance", "Musical", "Drama"]},
    {"title": "The Dark Knight", "genres": ["Action", "Thriller", "Crime"]},
    {"title": "Titanic", "genres": ["Romance", "Drama", "Disaster"]}
]

# User preferences
user_fav_genres = ["Romance", "Drama"]

# Recommend movies based on genre matching
def get_recommendations(user_genres, movies):
    scores = []
    for movie in movies:
        score = len(set(user_genres) & set(movie['genres']))
        scores.append((movie['title'], score))
    return sorted(scores, key=lambda x: x[1], reverse=True)

# Get top recommendations
recommendations = get_recommendations(user_fav_genres, movies)

# Prepare data for the graph
movie_titles = [title for title, score in recommendations]
match_scores = [score for title, score in recommendations]

# Plot the graph
plt.figure(figsize=(10, 6))
bars = plt.bar(movie_titles, match_scores, color='skyblue')
plt.xlabel("Movies")
plt.ylabel("Match Score")
plt.title("AI-Based Movie Recommendation Graph")
plt.xticks(rotation=45)
plt.ylim(0, max(match_scores) + 1)

# Show scores on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.1, int(yval), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
